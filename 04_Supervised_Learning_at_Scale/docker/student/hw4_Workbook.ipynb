# Question 1: Opimization Theory 

As you know from w207, Gradient Descent is an iterative process that seeks to find the optimal parameters for a model given a particular training data set. It does this by using the vector of partial derivatives of a loss function to strategically update parameters in a way that will reduce the loss. * __b) short response:__ Explain the relationship between problem domain space and model parameter space in the context of Gradient Descent. In practice, why can't we find the optimal model by simply looking at the error surface in model parameter space?

* __c) short response:__ In the context of Gradient Descent, what is the 'learning rate' and what are the tradeoffs associated with setting this hyperparameter?

* __d) BONUS:__ In the context of OLS, what do we mean by a 'closed form solution' and why is it not scalable?

### Q1 Student Answers:
> __a)__  
> The first-order condition of convexity is: A real-valued differentiable function is convex iff $f(x) \ge f(y) + \nabla f(y)^T (x-y)$, for all $x, y \in \mathbb{R}^n$. This concludes the function is globally above the tangent at $y$.  The first-order condition indicates that the maximum or minimum of an unconstrained function must occur at a point where the first derivative = 0.      
> The second-order condition of convexity is: A real-valued twice-differentiable function is convex iff $\nabla^2 f(x) \succeq 0$ for all $x \in \mathbb{R}^n$.  This concludes the function is flat or curved upward in every direction. The second-order condition determines if the zero value of the first derivative is a minimum or maximum.  It does this by taking the second derivative and evaluating the point.  If it is positive, we can conclude we are at a minimum.  
> We care because gradient descent is a method for optimizing many machine-learning algorithms, which relies on finding the parameters that minimize some objective function.  Having a convex objective function ensures a minimum is a global minimum.   
  
>   
   
> __b)__ Both the problem domain space and the model parameter space convey model loss.  In domain space, loss is naturally a measure of how far off the data are from the model.  In contrast, the loss in model parameter space is intuitively interpreted as an indicator of how far off the current model is from the optimal model.   In practice, we do not look at the error space to find the optimal model because it is computationally challenging.  The error space is only available after loss for every possible combination of parameters is computed.  There are better methods for arriving at the optimal solution.  
  
> __c)__ The learning rate is a hyperparameter that controls how much to change the model in response to the estimated error each time the model weights are updated. Choosing too small of a learning rate may result in a long training process by requiring more training epochs; the training process may also get stuck. In contrast, a value too large can cause the model to converge too quickly, resulting in a sub-optimal set of weights or an unstable training process.

> __d)__  In ordinary least squares, the goal is to find the line or hyperplane that minimizes the error between the target variable and the predicted output over all samples $i$ in some dataset of size $n$.  Solving for the model parameters analytically by inverting matrices is known as the closed-form solution; represented as $W = (X^TX)^{-1}X^Ty$.  It is considered closed form because $W$ can be calculated at once, unlike gradent descent which updates step by step. Unfortunately, matrix inversion is not easily parallelized and can be costly in higher dimensions, which could require more storage than possible in available memory.

```python
# part d - write function to compute loss (FILL IN MISSING CODE BELOW)
def OLSLoss(dataRDD, W):
    """
    Compute mean squared error.
    Args:
        dataRDD - each record is a tuple of (features_array, y)
        W       - (array) model coefficients with bias at index 0
    """
    augmentedData = dataRDD.map(lambda rec: (np.append([1.0], rec[0]), rec[1]))
    ################## YOUR CODE HERE ##################
    loss = None
    loss = augmentedData.map(lambda rec: (W.dot(rec[0].T) - rec[1])**2) \
                        .reduce(lambda a, b: a + b) / augmentedData.count()
    ################## (END) YOUR CODE ##################
    return loss
```

### Q4 Student Answers:
> __a)__ Gradient formula: $\nabla_{\theta} f(\theta) = \frac{2}{n} \sum_{i=1}^{n}[\theta^T \cdot x'_i-y_i] \cdot x'_i$  
The above equation shows that the phrase "the mean of the data weighted by the errors" relates to the gradient function by breaking it into two parts.  The mean of the data is expressed by $\frac{1}{n}$, while the error is expressed by $\sum_{i=1}^{n}[\theta^T \cdot x'_i-y_i]$

> __b)__ Everything to the right of the summation can be calculated in parallel.  The summation aggregation as well as the final multiplication of $\frac{2}{n}$ must be computed after the reduce.  The payload should contain the number of rows (records) to take the mean at the end.  Furthermore, utilizing combiners will sum local results.  The process is as follows:  First, chunk the input, calculate the formula inside the summation, and combine (sum) local results, then emit from mappers.  Next, a reduce job will sum the mapper outputs.  Lastly, the results are sent to the master node where the output is aggregated and multiplied by $\frac{2}{n}$ to compute the gradient.

> __c)__ _complete the coding portions of this question before answering d & e_ 

> __d)__ The results are not converging; the loss increases from step to step because the independent variables are not scaled appropriately.  Normalizing the features could help alleviate the issue by restricting the magnitude of the parameters to some constrained region.

The graphs plotted in Question 2 display features with very different value ranges (total sulfur dioxide ranges 0-450, citric acid ranges from 0-0.7, etc.); therefore, constraining them through normalization is recommended.
```python
# part c - function to perform a single GD step
def GDUpdate(dataRDD, W, learningRate = 0.1):
    """
    Perform one OLS gradient descent step/update.
    Args:
        dataRDD - records are tuples of (features_array, y)
        W       - (array) model coefficients with bias at index 0
    Returns:
        new_model - (array) updated coefficients, bias at index 0
    """
    # add a bias 'feature' of 1 at index 0
    augmentedData = dataRDD.map(lambda x: (np.append([1.0], x[0]), x[1])).cache()
    
    ################## YOUR CODE HERE ################# 
    grad = augmentedData.map(lambda X: (np.dot(X[0],W) - X[1]) * X[0]).mean() * 2
    new_model = W - np.multiply(grad, learningRate)
    ################## (END) YOUR CODE ################# 
   
    return new_model
```

```python
# part b - helper function to emit residuals (RUN THIS CELL AS IS)
def kResiduals(dataPoint, models, splitNum):
    """
    Compute the (squared) residuals for a data point given k different models.
    Note that points from the k-th split are part of the test set for model number k
    and part of the training set for all other models. We'll emit a key to track this.
    Args:
        dataPoint - tuple of (features_array, y)
        models    - list of arrays representing model weights (bias at index 0)
    Returns:
        (stringFormattedKey, squared_error)
    """
    # augment the data point with a bias term at index 0
    X = np.append([1.0], dataPoint[0])
    y = dataPoint[1]
    # emit squared residuals for each model
    for modelNum, W in enumerate(models):
        if modelNum == splitNum:
            yield(f"{modelNum}-test", (W.dot(X) - y)**2)
        else:
            yield(f"{modelNum}-train", (W.dot(X) - y)**2)
```

```python
# part c - fill in the missing code below
def CVLoss(dataSplits, models):
    """
    Compute the k-fold cross-validated test and train loss.
    Args:
        dataSplits - list of RDDs corresponding to the k test splits.
        models     - list of k arrays representing model weights (bias at index 0)
    Returns: 
        tuple of floats: (training_loss, test_loss)
    """
    # compute k residuals for each dataPoint (one for each model)
    partialLossRDD = sc.parallelize([])
    for splitNum, splitRDD in enumerate(dataSplits):
        residuals = splitRDD.flatMap(lambda x: kResiduals(x, models, splitNum))
        partialLossRDD = sc.union([partialLossRDD, residuals])
    
    ################ YOUR CODE HERE #################        
    loss = partialLossRDD.map(lambda rec: (rec[0], (rec[1], 1))) \
                         .reduceByKey(lambda a,b: (a[0] + b[0], a[1] + b[1])) \
                         .mapValues(lambda v: v[0]/v[1]) \
                         .collect()    
    ################ (END) YOUR CODE ################# 
    
    test_loss = np.mean([x[1] for x in loss if x[0].split('-')[1] == 'test'])
    training_loss = np.mean([x[1] for x in loss if x[0].split('-')[1] == 'train'])
    return training_loss, test_loss
```

```python
# part d - helper function RUN THIS CELL AS IS
def partialGradients(splitNum, dataPoint, models):
    """
    Emit partial gradient for this data point for each model.
    NOTE: a data point from split-number k is in the test set for 
    model-k so we don't compute a partial gradient for that model.
    """
    # augment the data point
    X = np.append([1.0], dataPoint[0])
    y = dataPoint[1]
    # emit partial gradients for each model with a counter for averaging later
    for modelNum, W in enumerate(models):
        if modelNum != splitNum:
            yield (modelNum, [(W.dot(X) - y)*X, 1])
```

```python
# part d - perform GD updates for all k models (FILL IN MISSING CODE BELOW)
def CVUpdate(dataSplits, models, learningRate = 0.1):
    """
    Compute gradients for k models given k corresponding dataSplits.
    NOTE: the training set for model-k is all records EXCEPT those in the k-th split.
    """
    # compute partial gradient k-1 times for each fold
    partialsRDD = sc.parallelize([])
    for splitNum, splitRDD in enumerate(dataSplits):
        thisFoldPartialGrads = splitRDD.flatMap(lambda x: partialGradients(splitNum, x, models))
        partialsRDD = sc.union([partialsRDD, thisFoldPartialGrads])

    # compute gradients by taking the average partialGrad for each fold
    gradients = partialsRDD.reduceByKey(lambda a, b: (a[0] + b[0], a[1] + b[1]))\
                           .mapValues(lambda x: x[0]/x[1])\
                           .map(lambda x: x[1])\
                           .collect()
    
    # update all k models & return them in a list
    new_models = []
    for W, grad in zip(models, gradients):
        ############# YOUR CODE HERE ############
        update = np.multiply(grad,learningRate)
        new_model = W - update
        new_models.append(new_model)
        ############# (END) YOUR CODE ###########
    return new_models
```

```python
# part e - RUN THIS CELL AS IS
def GradientDescent_withCV(dataSplits, wInit, learningRate=0.1, nSteps = 5, verbose = False):
    """
    Train k models in parallel and track cross validated test/train loss.
    Returns:
        train_hist - (list) of floats
        test_hist - (list) of floats
        model_hist - (list) of arrays representing model coefficients (bias at index 0)
    """
    # broadcast initial models (one for each fold)
    bModels = sc.broadcast([wInit] * len(dataSplits))
    
    
    # initialize lists to track performance
    train_loss_0, test_loss_0 = CVLoss(dataSplits, bModels.value)
    train_hist, test_hist, model_hist = [train_loss_0], [test_loss_0], [wInit]
    
    # perform k gradient updates at a time (one for each fold)
    start = time.time()
    for step in range(nSteps):
        new_models = CVUpdate(dataSplits, bModels.value, learningRate)
           
        bModels = sc.broadcast(new_models)

        # log progress
        train_loss, test_loss = CVLoss(dataSplits, bModels.value)
        train_hist.append(train_loss)
        test_hist.append(test_loss)
        model_hist.append(new_models[0])
        
        if verbose:
            print("-------------------")
            print(f"STEP {step}: ")
            print(f"model 1: {[round(w,4) for w in new_models[0]]}")
            print(f" train loss: {round(train_loss,4)}")
            print(f" test loss: {round(test_loss,4)}")
            
    print(f"\n... trained {nSteps} iterations in {time.time() - start} seconds")
    return train_hist, test_hist, model_hist
```

# Question 7: Regularization.

Our goal, as always, is to build a linear model that will extend well to unseen data. Chosing the right combination of features to optimize generalizability can be extremely computationally costly given that there are \\(2^{p}\\) potential models that can be built from \\(p\\) features. Traditional methods like forward selection would involve iteratively testing these options to asses which combinations of features achieve a statistically significant prediction.

Ridge Regression and Lasso Regression are two popular alternatives to OLS, which enable us to train generalizable models without the trouble of forward selection and/or manual feature selection.  Both methods take advantage of the bias-variance tradeoff by _shrinking_ the model coefficients towards 0 which reduces the variance of our model with little increase in bias. In practice this 'shrinkage' is achieved by adding a penalty (a.k.a. 'regularization') term to the means squared error loss function. In this question you will implement Gradient Descent with ridge and lasso regularization.

> __a)__ Ridge regularization loss function: $f(w,b) = \frac{1}{n} \sum_{i=1}^n [(w \cdot x_i+b)-y_i]^2 + \lambda \sum_{j=1}^m w_j^2$  
  
> Ridge regularization gradient function: $\nabla_\theta f(\theta) = \frac{2}{n} \sum_{i=1}^n [\theta^T \cdot x'_i - y_i] \cdot x'_i + \frac{\lambda}{n}(\theta)$  
  
> Ridge regularization adds a penalty to the gradient descent loss function.  Calculating the penalty will add a step to the gradient descent algorithm. Note that hyperparameter $\alpha$ defines the penalty strength on the loss. 
  

> __b)__ Lasso regularistion loss function: $f(w,b) = \frac{1}{n}\sum_{i=1}^n [(w \cdot x_i + b) - y_i]^2 + \lambda \sum_{j=1}^m |w_j|$  
  
> The derivative of the Lasso cost function does not have a closed-form solution which means the regularization parameter will be approximated by np.sign.  

> The lasso gradient descent update differs from the ridge regularization because it will make some coefficients zero, while the ridge regularization will not.

> __d)__ This does match my expectation.  Overall the regularization seems to improve the model's predictions.  I expected the ridge regression to look similar to the OLS regression, which it does.  A difference in the lasso regression is that over iterations, the loss oscillates due to model coefficients fluctuating from zero to non-zero values.

```python
# part c - gradient descent with regularization
def GDUpdate_wReg(dataRDD, W, learningRate = 0.1, regType = None, regParam = 0.1):
    """
    Perform one gradient descent step/update with ridge or lasso regularization.
    Args:
        dataRDD - tuple of (features_array, y)
        W       - (array) model coefficients with bias at index 0
        learningRate - (float) defaults to 0.1
        regType - (str) 'ridge' or 'lasso', defaults to None
        regParam - (float) regularization term coefficient
    Returns:
        model   - (array) updated coefficients, bias still at index 0
    """
    # augmented data
    augmentedData = dataRDD.map(lambda x: (np.append([1.0], x[0]), x[1]))
    
    #################### YOUR CODE HERE ###################      
    # compute gradient step
    grad = augmentedData.map(lambda row: (np.dot(row[0],W) - row[1]) * row[0]).mean() * 2
    
    # check regType and update model
    if regType == 'ridge':
        reg_term = regParam * 2 * W[1:]
        new_model = W - np.multiply(grad + np.append([0], reg_term), learningRate)

    elif regType == 'lasso':
        reg_term = regParam * np.sign(W[1:])
        new_model = W - np.multiply(grad + np.append([0], reg_term), learningRate)
    else:
        new_model = W - grad
    
    ################## (END) YOUR CODE ####################
    return new_model
```

```python
# part d - plotting function (RUN THIS CELL AS IS)
def plotCoeffs(models, featureNames, title):
    """
    Helper Function to show how coefficients change as we train.
    """
    fig, ax = plt.subplots(figsize = (15,8))
    X = list(range(len(models)))
    for data, name in zip(models.T, featureNames):
        if name == "Bias":
            continue
        ax.plot(X, data, label=name)
    ax.plot(X,[0]*len(X), 'k--')
    plt.title(title)
    plt.legend()
    plt.show()
```
